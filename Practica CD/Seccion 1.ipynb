#### SELECT

La instrucción SELECT en SQL se utiliza para recuperar y consultar datos de una base de datos

```python
%%sql
SELECT * FROM layoffs LIMIT 5;
```

```python
%%sql 
SELECT company, total_laid_off, percentage_laid_off, industry
FROM layoffs
GROUP BY industry, company, total_laid_off, percentage_laid_off
ORDER BY total_laid_off DESC
LIMIT 5;
```

#### WHERE
La cláusula WHERE en SQL se utiliza para filtrar los datos de una tabla. Permite especificar condiciones estrictas de modo que solo las filas que cumplan con dicho criterio sean devueltas en la consulta

#### AND | OR | NOT
Los operadores lógicos en SQL (AND, OR, NOT) permiten combinar múltiples condiciones en las cláusulas WHERE o HAVING para filtrar datos con precisión. Evalúan expresiones booleanas devolviendo TRUE, FALSE o NULL. Son fundamentales para realizar consultas complejas y segmentar información en bases de datos.

```python
%%sql
SELECT company, industry, location, total_laid_off
FROM layoffs
WHERE industry = 'Media'
GROUP BY location, company, industry
ORDER BY total_laid_off DESC
LIMIT 5
;
```

```python
%%sql
SELECT industry,total_laid_off, country
FROM layoffs
WHERE total_laid_off > 1000 AND country != 'United States'
GROUP BY industry, country, total_laid_off
LIMIT 10
;
```

```python
%%sql
SELECT industry, company, SUM(total_laid_off) AS total_laid_off
FROM layoffs
GROUP BY industry, company
ORDER BY total_laid_off DESC
LIMIT 10;
```

#### LIKE
El operador LIKE en SQL se utiliza en la cláusula WHERE para buscar un patrón específico dentro de una columna de texto. Permite realizar búsquedas flexibles y coincidencias parciales usando comodines, a diferencia del operador = que exige una coincidencia exacta.

```python
%%sql
SELECT *
FROM layoffs
WHERE company LIKE '%tla%'
;
```

#### GROUP BY
La cláusula GROUP BY en SQL agrupa filas con valores idénticos en columnas específicas para generar filas de resumen. Se utiliza junto con funciones de agregación (como SUM, COUNT, AVG, MAX o MIN) para realizar cálculos o estadísticas sobre cada grupo de datos.

#### ORDER BY
En SQL, la cláusula ORDER BY se utiliza para ordenar los resultados de una consulta de forma ascendente o descendente. Por defecto, ordena de menor a mayor (A-Z), pero puedes personalizarla fácilmente para adaptarla a cualquier necesidad.

```python
%%sql
SELECT industry, AVG(total_laid_off) AS avg_laid_off
FROM layoffs
GROUP BY industry
ORDER BY avg_laid_off DESC
LIMIT 5
;
```

```python
%%sql
SELECT total_laid_off, industry
FROM layoffs
GROUP BY industry
ORDER BY total_laid_off DESC
;
```

```python
%%sql
SELECT industry, AVG(total_laid_off) AS avg_laid_off, MAX(total_laid_off), MIN(total_laid_off)
FROM layoffs
GROUP BY industry
ORDER BY avg_laid_off DESC
LIMIT 5
;
```

```python
%%sql
SELECT company, industry, SUM(total_laid_off) AS total_laid_off, COUNT(*) AS num_events
FROM layoffs
GROUP BY company, industry
LIMIT 10
;
```

```python
%%sql
SELECT *
FROM layoffs
WHERE company = '2TM'
;
```

#### WHERE vs HAVING
WHERE filtra filas individuales antes de agrupar los datos. No admite funciones de agregación (como \(COUNT\), \(SUM\), o \(AVG\)). HAVING filtra los grupos de filas después de agrupar y se utiliza específicamente con resultados agregados

```python
%%sql
SELECT company, industry, total_laid_off
FROM layoffs
GROUP BY industry, company
HAVING total_laid_off > 1000
ORDER BY total_laid_off DESC
;
```
